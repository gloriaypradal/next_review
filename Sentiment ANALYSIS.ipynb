I am going to work with Vader sentiment analysis, is a simple lexicon and rule-based model for general setiment analysis. This analysis results gives not only positive and negative. but also how much of each polarity you find in the text that we are analyzing, also is easier and the code is very straight forward. It does not require trained data, and because of the amount of sentences that we have in the data set, it is more convinient to work with data without training

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

I found this example on: https://medium.com/swlh/simple-sentiment-analysis-for-nlp-beginners-and-everyone-else-using-vader-and-textblob-728da3dbe33d 
and i coied here for me to check how it actually works

```python
analyzer= SentimentIntensityAnalyzer()
sentence = "Today is a SUPER good day"
analyzer.polarity_scores(sentence)
#{'neg': 0.0, 'neu': 0.494, 'pos': 0.506, 'compound': 0.6249}
```

To add the scores that Vader setiment analysis gives me, I am going to add the results to the data frame

```python
df['Compound'] = [analyzer.polarity_scores(a)['compound'] for a in df.comments]
df['Negative'] = [analyzer.polarity_scores(a)['neg'] for a in df.comments]
df['Positive'] = [analyzer.polarity_scores(a)['pos'] for a in df.comments]
df['Neutral'] = [analyzer.polarity_scores(a)['neu'] for a in df.comments]
```

```python
df
```

I am going to use the compound score to classify if the comment is positive, neutral or negative because the compound is "a normalized weighted composite computed by summing the valence score of each word in the lexicon, adjusted according to the rules, and then normalized to be between -1 and 1" (Kleppen. https://medium.com/swlh/simple-sentiment-analysis-for-nlp-beginners-and-everyone-else-using-vader-and-textblob-728da3dbe33d), for this assignment we are only asked to classify negative and positive, but in my opinion if we want to go to a little bit further saying that a comment is "kinda of poitive" or "mild negative" this score (compound) would be usefull
Stablishing the criteria that if the Compund score is positive, the comment is positive, if it's == 0 is neutral and if is negative the comment is negative, we have:

```python
for a in df.Compound:
    df['Analysis'] = df.Compound.apply(lambda a: 'Positive' if a > 0 else ('Neutral' if a == 0 else 'Negative'))
```

```python
df
```

```python
for a in range(len(df.comments)):
    print('Comment:\n', df.comments[a])
    print('Sentiment Analisis:', df.Analysis[a])
```
